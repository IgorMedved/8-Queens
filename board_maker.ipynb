```python
import jupyter_loader
jupyter_loader.load()
import Queens
import pygame
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QSlider, QVBoxLayout, QLabel
from PyQt5.QtGui import QImage, QPainter
from PyQt5.QtCore import QObject, Qt, pyqtSignal
from matplotlib import image as mpimg
import numpy as np
```

```python
#UI for the solution
#the ideo on implementation of embedding pygame into PyQt is taken from 
#https://stackoverflow.com/questions/38280057/how-to-integrate-pygame-and-pyqt4

class ImageWidget(QWidget):
    def __init__(self,surface,parent=None):
        super().__init__()
        self.updateImage(surface)
        self.setMinimumSize(250,250)
        

    def paintEvent(self,event):
        qp=QPainter()
        qp.begin(self)
        qp.drawImage(0,0,self.image)
        qp.end()
        
    def updateImage(self, surface):

        w=surface.get_width()
        h=surface.get_height()
        self.data=surface.get_buffer().raw
       
        self.image=QImage(self.data,w,h, QImage.Format_RGB32)


class QtInterface(QWidget):
    def __init__(self, parent = None):
        super(QtInterface, self).__init__(parent)
        self.rows, self.cols, self.solutionsSet = Queens.init()
        self.pygameInit()
        self.initUI()
        
    def initUI(self):
        # board size 300 x 350
        self.setGeometry(250,250,300,350)
        self.setWindowTitle ('8 Queens Problem')
        self.updateFrame(self.solutionsSet[0])
        
        self.board = ImageWidget(self.gameDisplay)
        #self.board.setGeometry(0,0,330,330)
        vbox = QVBoxLayout()
        #vbox.addStretch(1)
       
        sld = QSlider (Qt.Horizontal, self)
        sld.setFocusPolicy(Qt.NoFocus)
        sld.setGeometry(10,300,100,30)
        
        sld.valueChanged[int].connect(self.changeValue)
       
        sld.setMaximum(len(self.solutionsSet)-1)
        vbox.addWidget(QLabel('Solution:'))
        vbox.addWidget(self.board)
        vbox.addWidget(QLabel('Select solution with slider'))
        vbox.addWidget(sld)
        
        self.setLayout(vbox)
        self.show()
        
    def changeValue(self, value):

        self.updateFrame(self.solutionsSet[value])
        self.board.updateImage(self.gameDisplay)        
        self.board.repaint()
    
    def pygameInit(self):
        pygame.init()
        screen_width, screen_height = 250,250
        self.gameDisplay = pygame.Surface((screen_width, screen_height))
        
        self.queenb = pygame.image.load('black_transp.png')
        
    def updateFrame(self, solution):
        chocolate = (210,105,30,255)
        silk = (255,248,220,255)
        tan = (210,180,140, 255)

        size = (30,30)
        
        def plotSolution (solution):
            for box in solution:
                j = self.rows.index(box[1])
                i = self.cols.index(box[0])
               # print (i, j)
                square = i*size[0]+13, (7-j)*size[1]+5
                self.gameDisplay.blit(self.queenb, square)
        

    
        self.gameDisplay.fill(tan)
        for i in range(8):
            for j in range(8):
                square = i*30+5, j*30+5
                if ((i+j)%2 == 0):
                    self.gameDisplay.fill(silk, (square, size))
                else:
                    self.gameDisplay.fill(chocolate, (square, size))
            
            
        plotSolution(solution)
        
        
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = QtInterface()
    app.exec_()
    pygame.quit()
    exit
```
